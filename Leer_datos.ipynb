```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("data.csv")
```

```python
df.head(1)
```

```python
df["categoria"].value_counts()
```

```python
plt.figure(figsize=(8, 5))
sns.countplot(data=df,
              x="categoria",
              order=df["categoria"].value_counts().index
             )

plt.title("Cantidad de ventas por categoria")
plt.show()
```

```python
df["dia_semana"] = pd.to_datetime(df["fecha"]).dt.day_name()

plt.figure(figsize=(8, 5))
sns.countplot(data=df,
              x="dia_semana",
              order=["Monday","Tuesday","Wednesday","Thursday","Friday","Saturday","Sunday"]
             )

plt.title("Cantidad de ventas por dia de la semana")
plt.show()
```

```python
plt.figure(figsize=(12, 5))

sns.histplot(df["hora"], bins=12,kde=True)

plt.title("Cantidad de ventas por hora del dia")
plt.show()
```
